```python
class Solution:
    def sort012(self, arr):
        # code here
        length = len(arr)
        low = 0
        high = length-1
        mid = 0
        while mid <= high:
            if arr[mid]==0:
                arr[low], arr[mid] = arr[mid], arr[low]
                mid+=1
                low+=1
            elif arr[mid] == 1:
                mid+=1
            elif arr[mid] ==2:
                arr[mid], arr[high] = arr[high],arr[mid]
                high-=1
        return arr
```

```python
class Solution:
    def binarysearch(self, arr, k):
        # Code Here
        test = lambda x:True if x==k else False
        for i in range(len(arr)):
            if test(arr[i]):
                return i
        return -1
```
