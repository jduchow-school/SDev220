```python
secret=6
guess=7
if guess == secret:
    print("just right")
elif guess > secret:
    print("too high")
else:
    print("too low")
```

```python
small = True
green = False
if small:
    if green:
        print("this is a pea")
    else:
        print("this is a cherry")
else:
    if green:
        print("this is a watermelon")
    else:
        print("this is a pumpkin")
    
```

```python
guess_me = 7
number = 1
while True:
    if number > guess_me:
        print("oops")
        break
    elif number == guess_me:
        print("found it!")
        break
    number+=1
    print("too low")
```

```python
guess_me = 5
for number in range(10):
    if number > guess_me:
        print("oops")
        break
    elif number == guess_me:
        print("found it!")
        break
    print("too low")
```
