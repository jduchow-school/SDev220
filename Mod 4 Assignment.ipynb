```python
import csv
import sqlite3 as sq
db = sq.connect('books.db')
ins_str = 'insert into book values(?, ?, ?)'
curs = db.cursor()
with open('books2.csv','rt')as infile:
    books = csv.DictReader(infile)
    for book in books:
        curs.execute(ins_str, (book['title'], book['author'], book['year']))
db.commit()
db.close()
```

```python
import sqlalchemy as sqla
from sqlalchemy import text

# Create engine
conn = sqla.create_engine('sqlite:///books.db')

# Create a connection from the engine
with conn.connect() as connection:
    # Execute SQL query using the connection
    # Use text() to convert the SQL string to an executable object
    sql = text('select title from book order by title asc')
    rows = connection.execute(sql)
    
    # Iterate through results
    for row in rows:
        print(row)
```
