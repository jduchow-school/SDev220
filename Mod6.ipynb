```python
#prints date on to today.txt
from datetime import date
now = date.today()
nowfrmt = now.isoformat()
with open('today.txt','wt') as fout:
    print(nowfrmt, file = fout)
```

```python
#read today.txt and convert into a string
with open('today.txt','rt') as fout:
    today_string = fout.read()
print(today_string)
```

```python
#parse today_string
from datetime import datetime
fmt = "%Y-%m-%d"
datetime.strptime(today_string.strip(), fmt)
```

```python
#use multiprocessing to create 3 processes which wait a random nuber of seconds and then prints the current time
import multiprocessing

def timeNow(seconds):
    import time
    from datetime import datetime
    time.sleep(seconds)
    print(datetime.now())

if __name__ == "__main__":
    import random
    for i in range(3):
        seconds = random.random()
        p = multiprocessing.Process(target = timeNow, args =(seconds,))
        p.start()
```
